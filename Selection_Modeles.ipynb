```python
import pandas as pd
import numpy as np
import joblib

# 1. Chargement du Mod√®le
MODEL_FILENAME = "modele_charges_final.joblib"
final_model = joblib.load(MODEL_FILENAME)
print(f"‚úÖ Mod√®le charg√© : {MODEL_FILENAME}")

# --- 2. D√©finition des Donn√©es d'Entr√©e (Conversion des Types et FE) ---

# Votre ligne de donn√©es : [age, bmi, children, region (1), sex_male (0), smoker_yes (1)]
AGE = 29.0
BMI = 27.5
CHILDREN = 2.0
REGION_CODE = 1.0
SEX_MALE = False  # Conversion de 0 √† False (Bool√©en)
SMOKER_YES = True # Conversion de 1 √† True (Bool√©en)

# Calcul de la Feature d'Interaction (Obligatoire pour votre mod√®le)
# L'interaction est Age * Smoker_yes (num√©rique)
AGE_SMOKER_INTERACTION = AGE * (1 if SMOKER_YES else 0)

# 3. Cr√©ation de la DataFrame avec la structure et les types EXACTS
new_data_final = pd.DataFrame([{
    'age': AGE,
    'bmi': BMI,
    'children': CHILDREN,
    'region': REGION_CODE,
    'sex_male': SEX_MALE,
    'smoker_yes': SMOKER_YES,
    'age_smoker_interaction': AGE_SMOKER_INTERACTION
}])

# 4. D√©finition et application de l'Ordre Final (Le mod√®le l'exige)
COLUMNS_ORDER_REQUIRED = [
    'age', 'bmi', 'children', 'region', 'sex_male', 'smoker_yes', 'age_smoker_interaction'
]
new_data_final = new_data_final[COLUMNS_ORDER_REQUIRED]

# --- 5. Pr√©diction et Transformation Inverse ---

# Pr√©diction
pred_log = final_model.predict(new_data_final)[0]

# Transformation inverse (exp(x) - 1)
pred_charges = np.expm1(pred_log)

# Affichage du r√©sultat
print("\n--- R√©sultat de la Pr√©diction ---")
print(f"Caract√©ristiques align√©es : {new_data_final.to_dict(orient='records')[0]}")
print(f"üí∞ Charges M√©dicales Pr√©dites : **${pred_charges:.2f}**")
```
