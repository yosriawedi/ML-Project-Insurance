```python
import pandas as pd
import numpy as np
import joblib

# 1. Chargement du Modèle
MODEL_FILENAME = "modele_charges_final.joblib"
final_model = joblib.load(MODEL_FILENAME)
print(f"✅ Modèle chargé : {MODEL_FILENAME}")

# --- 2. Définition des Données d'Entrée (Conversion des Types et FE) ---

# Votre ligne de données : [age, bmi, children, region (1), sex_male (0), smoker_yes (1)]
AGE = 29.0
BMI = 27.5
CHILDREN = 2.0
REGION_CODE = 1.0
SEX_MALE = False  # Conversion de 0 à False (Booléen)
SMOKER_YES = True # Conversion de 1 à True (Booléen)

# Calcul de la Feature d'Interaction (Obligatoire pour votre modèle)
# L'interaction est Age * Smoker_yes (numérique)
AGE_SMOKER_INTERACTION = AGE * (1 if SMOKER_YES else 0)

# 3. Création de la DataFrame avec la structure et les types EXACTS
new_data_final = pd.DataFrame([{
    'age': AGE,
    'bmi': BMI,
    'children': CHILDREN,
    'region': REGION_CODE,
    'sex_male': SEX_MALE,
    'smoker_yes': SMOKER_YES,
    'age_smoker_interaction': AGE_SMOKER_INTERACTION
}])

# 4. Définition et application de l'Ordre Final (Le modèle l'exige)
COLUMNS_ORDER_REQUIRED = [
    'age', 'bmi', 'children', 'region', 'sex_male', 'smoker_yes', 'age_smoker_interaction'
]
new_data_final = new_data_final[COLUMNS_ORDER_REQUIRED]

# --- 5. Prédiction et Transformation Inverse ---

# Prédiction
pred_log = final_model.predict(new_data_final)[0]

# Transformation inverse (exp(x) - 1)
pred_charges = np.expm1(pred_log)

# Affichage du résultat
print("\n--- Résultat de la Prédiction ---")
print(f"Caractéristiques alignées : {new_data_final.to_dict(orient='records')[0]}")
print(f"💰 Charges Médicales Prédites : **${pred_charges:.2f}**")
```
